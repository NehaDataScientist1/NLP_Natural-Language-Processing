```python
import nltk

para= """Artificial Intelligence (AI) is a rapidly evolving field that encompasses a wide range of technologies aimed at mimicking human intelligence. From natural language processing to computer vision, AI applications span various domains, including healthcare, finance, and entertainment. With advancements in machine learning algorithms and computational power, AI systems are becoming increasingly capable of performing complex tasks autonomously.

The potential of AI to revolutionize industries and streamline processes is immense, promising greater efficiency, productivity, and innovation. However, concerns about the ethical implications of AI, such as algorithmic bias and privacy breaches, underscore the need for responsible development and deployment. Striking a balance between harnessing the benefits of AI and addressing its societal impacts is essential for shaping a future where AI serves humanity best interests."""
```

```python
para
```

```python
sentences= nltk.sent_tokenize(para)
sentences
```

```python
words= nltk.word_tokenize(para)
words
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
stemmer=PorterStemmer()
```

### Code to remove all stop words from  Paragraph:
#### 1. using PorterStemmer

```python
sentence1= sentences.copy()
sentence1
```

```python
for i in range(len(sentence1)):
    words= nltk.word_tokenize(sentence1[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentence1[i]= ' ' .join(words)     
sentence1
```

#### 2. using LancasterStemmer:

```python
sentence2= sentences.copy()
sentence2
from nltk.stem import LancasterStemmer
from nltk.corpus import stopwords
Lan_stemmer=LancasterStemmer()
for i in range(len(sentence2)):
    words= nltk.word_tokenize(sentence2[i])
    words = [Lan_stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentence2[i]= ' ' .join(words)     
sentence2
```

#### 3. using Lemmatization:

```python
sentence3= sentences.copy()
sentence3
from nltk.stem import wordnet
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

lemmatization= WordNetLemmatizer()

for i in range(len(sentence3)):
    words= nltk.word_tokenize(sentence3[i])
    words = [lemmatization.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    sentence3[i]= ' ' .join(words)     
sentence3
```

# Name Entity Recognition (NER):


```python
from nltk import ne_chunk
NE_sent ='The US president stays in whitehouse'
NE_tokens = word_tokenize(NE_sent)
NE_tokens
```

```python
NE_tags = nltk.pos_tag(NE_tokens)
NE_tags
```

### NER

```python
from nltk import ne_chunk
NE_NER = ne_chunk (NE_tags)
print(NE_NER)
```

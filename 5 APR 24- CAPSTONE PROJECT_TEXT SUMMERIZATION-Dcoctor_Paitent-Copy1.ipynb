#### Calculate frequency of each and every word , how many times word is repeated in text:
code:

```python
word_frequencies={}
```

```python
type(word_frequencies)
```

```python
for word in doc:
    if word.text.lower() not in stopwords:
        if word.text.lower() not in punctuation:
            if word.text not in ('\n', 'Doctor', 'Patient'):
                if word.text not in word_frequencies.keys():
                    word_frequencies[word.text]=1
                else:
                    word_frequencies[word.text]+=1
```

```python
word_frequencies
```

```python
len(word_frequencies)
```

```python
max_frequency = max(word_frequencies.values())
max_frequency
```

#### To get normalised/weighted frequencies you should divide all frequnecies with max frequency:

```python
for word in word_frequencies.keys():
    word_frequencies[word] = word_frequencies[word]/max_frequency
```

```python
word_frequencies
```

#### Now we need to merge all tokens score in sentences to get senteces score:

```python
sentence_token = [sent for sent in doc.sents]
```

```python
len(sentence_token)
```

```python
sentence_token
```

#### Now we need to calculate the sentence score :
code:

```python
sentence_scores = {}
for sent in sentence_token: # 1st sentnecs in sentences toke where all text divided into individual sentnecs
    for word in sent:       # for each work token inthat 1 sentnece
        if word.text.lower() in word_frequencies.keys(): # if 1 word is in word_frequencies list
            if sent  not in sentence_scores.keys():       # also 1 word not in sentnece_score 
                sentence_scores[sent]= word_frequencies[word.text.lower()] # then add 1 word score in that words score count which was empty 
            else:
                sentence_scores[sent] += word_frequencies[word.text.lower()]  # if 1 word is in sentnece_score then add new score to that words score count

```

```python
sentence_scores
```

```python
len(sentence_scores)
```

### DataStructure libraray : **Heapq**

```python
from heapq import nlargest
```

```python
select_length = int (len(sentence_token)*0.3)
select_length
```

#### We will select max 7 sentences having max 7 sentnces score from all sentnecs:

```python
summary = nlargest(select_length,sentence_scores,key= sentence_scores.get)
summary
```

```python
sentence_scores
```

```python
final_summary= [word.text for word in summary]
final_summary
```

### Also we can create the wordcloud of Important words to quicl check for doctor:

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import re
```

```python
text
```

```python
import nltk

               
 #cleaing text

import re
from nltk.corpus import stopwords             
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

ps= PorterStemmer()
wordnet = WordNetLemmatizer()

# convert para to sentences
sentences = nltk.sent_tokenize(text)

corpus= []   # empty list


for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentences[i])
    review = review.lower()
    review = review.split()


    review = [wordnet.lemmatize(word) for word in review if not word in set (stopwords.words('english'))]


    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
modified_strings = []

for string in corpus:
    words = string.split()  # Split the string into words
    filtered_words = [word for word in words if word.lower() not in ['doctor', 'patient']]
    modified_string = ' '.join(filtered_words)  # Join the filtered words back into a string
    modified_strings.append(modified_string)

print(modified_strings)

# Join the strings in the list with a space between each string
result_string = ' '.join(modified_strings)

print(result_string)
```

```python
wc = WordCloud(width=700, height=700 ,margin = 0 ,colormap='cividis').generate(result_string)
plt.imshow(wc, interpolation='nearest' )
plt.axis('off')
plt.margins(x=0, y=0)
plt.show()
```

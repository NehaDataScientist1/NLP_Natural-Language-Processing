```python
text = """Doctor: Good afternoon. How are you feeling today?
Patient: I've been feeling a bit off lately. I've had some headaches and I'm feeling more tired than usual.
Doctor: I see. Have you been monitoring your blood pressure at home?
Patient: Yes, I have. It's been consistently high for the past week or so.
Doctor: I'm concerned about that. High blood pressure can lead to serious health issues if left untreated. Have you made any changes to your diet or lifestyle recently?
Patient: Not really, everything's been the same as usual.
Doctor: Alright. I'd like to take your blood pressure now and discuss some options for managing it. We may need to consider medication, along with lifestyle changes such as dietary adjustments and increasing physical activity.
Patient: I understand. I'm willing to do whatever it takes to get it under control.
Doctor: That's the right attitude to have. Let's start by checking your blood pressure and then we'll formulate a plan together.
Doctor: Your blood pressure is indeed high. I'm going to prescribe medication to help lower it, but it's important that you also make those lifestyle changes we discussed.
Patient: I'll definitely do that. Thank you, doctor, for taking this seriously.
Doctor: Of course. Managing your blood pressure is crucial for your overall health. We'll schedule a follow-up appointment to monitor your progress and adjust treatment if necessary.
Patient: Sounds good. I appreciate your help, doctor.
Doctor: You're welcome. Remember, if you experience any concerning symptoms or have questions, don't hesitate to reach out.
Patient: Will do. Thank you again, doctor.
Doctor: Take care, and we'll see you soon for your follow-up appointment. """
```

#### Calculate frequency of each and every word , how many times word is repeated in text:
code:

```python
word_frequencies={}
```

```python
type(word_frequencies)
```

```python
for word in doc:
    if word.text.lower() not in stopwords:
        if word.text.lower() not in punctuation:
            if word.text not in ('\n', 'Doctor', 'Patient'):
                if word.text not in word_frequencies.keys():
                    word_frequencies[word.text]=1
                else:
                    word_frequencies[word.text]+=1
```

```python
word_frequencies
```

```python
len(word_frequencies)
```

```python
max_frequency = max(word_frequencies.values())
max_frequency
```

#### To get normalised/weighted frequencies you should divide all frequnecies with max frequency:

```python
for word in word_frequencies.keys():
    word_frequencies[word] = word_frequencies[word]/max_frequency
```

```python
word_frequencies
```

#### Now we need to merge all tokens score in sentences to get senteces score:

```python
sentence_token = [sent for sent in doc.sents]
```

```python
len(sentence_token)
```

```python
sentence_token
```

#### Now we need to calculate the sentence score :
code:

```python
sentence_scores = {}
for sent in sentence_token: # 1st sentnecs in sentences toke where all text divided into individual sentnecs
    for word in sent:       # for each work token inthat 1 sentnece
        if word.text.lower() in word_frequencies.keys(): # if 1 word is in word_frequencies list
            if sent  not in sentence_scores.keys():       # also 1 word not in sentnece_score 
                sentence_scores[sent]= word_frequencies[word.text.lower()] # then add 1 word score in that words score count which was empty 
            else:
                sentence_scores[sent] += word_frequencies[word.text.lower()]  # if 1 word is in sentnece_score then add new score to that words score count

```

```python
sentence_scores
```

```python
len(sentence_scores)
```

### DataStructure libraray : **Heapq**

```python
from heapq import nlargest
```

```python
select_length = int (len(sentence_token)*0.3)
select_length
```

#### We will select max 7 sentences having max 7 sentnces score from all sentnecs:

```python
summary = nlargest(select_length,sentence_scores,key= sentence_scores.get)
summary
```

```python
final_summary= [word.text for word in summary]
final_summary
```
